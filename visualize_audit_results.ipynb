```python
import os
import random
import re

# 之前生成的匹配文件列表
INPUT_LIST_FILE = "matched_audit_files.txt"
# 图片所在的根目录 (需要根据您的实际挂载路径调整，这里假设是相对路径)
# 注意：Jupyter Notebook 运行时的路径可能与当前脚本不同，这里使用相对路径
IMG_BASE_DIR = "pico-banana-400k-subject_driven/openimages/pico_sam_output_ALL_20251206_032609"

def generate_jupyter_code(num_samples=5):
    if not os.path.exists(INPUT_LIST_FILE):
        print(f"Error: {INPUT_LIST_FILE} not found.")
        return

    with open(INPUT_LIST_FILE, 'r') as f:
        files = [line.strip() for line in f if line.strip()]

    if not files:
        print("No files found in list.")
        return

    # 随机抽取样本
    selected_files = random.sample(files, min(num_samples, len(files)))
    
    # 提取 item_id 并构建数据结构
    display_data = []
    for filepath in selected_files:
        # 从文件名提取 item_id，例如 item_1017_dino_audit.json -> 1017
        match = re.search(r"item_(\d+)_dino_audit\.json", filepath)
        if match:
            item_id = match.group(1)
            item_dir = f"item_{item_id}"
            
            # 构建图片路径
            img_add = os.path.join(IMG_BASE_DIR, item_dir, f"item_{item_id}_MASK_ADD.png")
            img_remove = os.path.join(IMG_BASE_DIR, item_dir, f"item_{item_id}_MASK_REMOVE.png")
            img_merged = os.path.join(IMG_BASE_DIR, item_dir, f"item_{item_id}_MERGED.png")
            
            display_data.append({
                "id": item_id,
                "add": img_add,
                "remove": img_remove,
                "merged": img_merged
            })

    # 生成 Jupyter 代码字符串
    jupyter_code = f"""
import matplotlib.pyplot as plt
from PIL import Image
import os

# 待展示的数据列表
data_to_show = {json.dumps(display_data, indent=4)}

def show_images(data_list):
    for item in data_list:
        item_id = item['id']
        paths = [item['add'], item['remove'], item['merged']]
        titles = [f"Item {{item_id}} - ADD", f"Item {{item_id}} - REMOVE", f"Item {{item_id}} - MERGED"]
        
        # 检查文件是否存在
        valid_imgs = []
        valid_titles = []
        for p, t in zip(paths, titles):
            if os.path.exists(p):
                valid_imgs.append(Image.open(p))
                valid_titles.append(t)
            else:
                print(f"Image not found: {{p}}")
        
        if not valid_imgs:
            continue
            
        # 绘图
        plt.figure(figsize=(15, 5))
        for i, (img, title) in enumerate(zip(valid_imgs, valid_titles)):
            plt.subplot(1, 3, i+1)
            plt.imshow(img)
            plt.title(title)
            plt.axis('off')
        plt.tight_layout()
        plt.show()

# 执行展示
print(f"Displaying {{len(data_to_show)}} random samples...")
show_images(data_to_show)
"""
    
    print("-" * 20 + " Copy below code to Jupyter Notebook " + "-" * 20)
    print(jupyter_code)
    print("-" * 60)

if __name__ == "__main__":
    import json
    generate_jupyter_code()
```
